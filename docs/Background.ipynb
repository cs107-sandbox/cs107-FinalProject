## 1. The Chain Rule
$y = f(g(h(x)))$
From inside to out side: 
$$w1 = h(x)$$
$$w2 = g(w1)$$
$$y = f(w2)$$
The derivative with respect to x: 
$$\frac{dy}{dx} = \frac{dy}{dw_{2}} \frac{dw_{2}}{dw_{1}} \frac{dw_{1}}{dx}$$
